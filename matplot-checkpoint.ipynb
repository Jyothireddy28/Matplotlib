```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated Test Scores for Three Classes
class_A = np.random.normal(75, 10, 100)  # Mean 75, Std 10
class_B = np.random.normal(70, 15, 100)  # Mean 70, Std 15
class_C = np.random.normal(80, 5, 100)   # Mean 80, Std 5

# Combine Data
data = [class_A, class_B, class_C]

# Plotting the Violin Plot
plt.figure(figsize=(8, 6))
plt.violinplot(data, showmeans=True, showmedians=True)

# Customization
plt.title('Distribution of Student Scores Across Classes')
plt.xticks([1, 2, 3], ['Class A', 'Class B', 'Class C'])
plt.ylabel('Scores')
plt.grid(True)

plt.show()
```
